### Year vs Released Columns
There seems to be a descripency in some areas between the 'year' and 'relased' columns. The expectation is for the years to correlate. 

To fix this, the year value from the 'reased' column will be used in the 'year' column



```python
movies.tail(20)
```

```python
movies['yearcorrect'] = movies['released'].str.extract(pat = '([0-9]{4})').astype(float)    
```

```python
movies.head()
```

```python
if movies['yearcorrect'].isna().any():
    movies['yearcorrect'] = movies['year']
    
   
```

```python
movies['yearcorrect'] = movies['yearcorrect'].astype(int)
```

```python
movies.head()
```

```python
movies.sort_values(by=['gross'], inplace=False, ascending=False)
```

### Filling in the missing values with Median values

```python
movies['budget'].fillna(movies['budget'].median(), inplace=True)
movies['gross'].fillna(movies['gross'].median(), inplace=True)
```

### Looking at correlations with non-numeric data(Strings)

The way that this will be done will be by converting the 'categories' into numbers

```python
movies_numerised = movies

for col_name in movies_numerised:
    if(movies_numerised[col_name].dtype == 'object'):
        movies_numerised[col_name] = movies_numerised[col_name].astype('category')
        movies_numerised[col_name] = movies_numerised[col_name].cat.codes
        
movies_numerised.head()
```

```python
correlation_matrix = movies_numerised.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix for the variables")
plt.xlabel("movie features")
plt.ylabel("movie features")
```

```python
correlation_mat = movies_numerised.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs.head()
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5] 
high_corr
```
