## Validation Tests for Date and time

```python
def test_date_column(date_series):
    """
    Test if the Date column contains only valid dates (in YYYY-MM-DD format).
    Returns True if all values are valid dates, False otherwise.
    """
    try:
        pd.to_datetime(date_series, format='%Y-%m-%d', errors='raise')
        return True
    except (ValueError, TypeError):
        return False

def test_time_column(time_series):
    """
    Test if the Time column contains only valid times (in HH:MM or HH:MM:SS format).
    Returns True if all values are valid times, False otherwise.
    """
    for time_value in time_series:
        try:
            # Try parsing time in both HH:MM:SS and HH:MM formats
            pd.to_datetime(time_value, format='%H:%M:%S', errors='raise')
        except (ValueError, TypeError):
            try:
                pd.to_datetime(time_value, format='%H:%M', errors='raise')
            except (ValueError, TypeError):
                return False
    return True


date_test_result = test_date_column(SpaceX_df['Date'])
time_test_result = test_time_column(SpaceX_df['Time'])

# Output test results
print(f"Date column valid: {date_test_result}")
print(f"Time column valid: {time_test_result}")
```

### Version, booster

```python
SpaceX_df[["Version, booster"]].sample(15)
```

```python
import re

def clean_version_booster(entry):
    # Standardize 'F9B5' to 'F9 B5' and 'F9FT' to 'F9 FT'
    entry = re.sub(r'F9B5', 'F9 B5', entry)
    entry = re.sub(r'F9FT', 'F9 FT', entry)

    # Ensure consistent space after Falcon versions like 'F9 B4', 'F9 B5'
    entry = re.sub(r'F9B', 'F9 B', entry)

    # Fix missing spaces between booster and version numbers (e.g., 'F9B5' to 'F9 B5')
    entry = re.sub(r'(?<=F9)(?=\d)', ' ', entry)

    # Standardize Falcon Heavy by removing only '(core)' and keeping the booster info
    entry = re.sub(r'Falcon Heavy (B\d+\sB\d+)\s*\(core\)', r'Falcon Heavy \1', entry)

    # Fix booster number formatting to remove extra spaces (e.g., 'B1051 .4' to 'B1051.4')
    entry = re.sub(r'(\bB\d+)\s*\.\s*(\d+)', r'\1.\2', entry)

    # Remove redundant 'F9 B5' after the main identifier
    entry = re.sub(r'(F9 B5 \w+\.\d+) F9 B5', r'\1', entry)
    entry = re.sub(r'(F9 B4 \w+\.\d+) F9 B4', r'\1', entry)
    entry = re.sub(r'(F9 FT \w+\.\d+) F9 FT', r'\1', entry)

    return entry


SpaceX_df['cleaned_version_booster'] = SpaceX_df['Version, booster'].apply(clean_version_booster)
SpaceX_df[['Version, booster', 'cleaned_version_booster']].sample(10)
```

```python
def extract_payload_info(payload):
    # Extract mission name (everything before ':' or '(')
    mission_name = re.split(r"[:(]", payload)[0].strip()

    # Updated regex to extract satellite count within parentheses
    satellite_count = re.findall(r"\((\d+)\s+satellites?\)", payload)
    satellite_count = int(satellite_count[0]) if satellite_count else None

    # Extract spacecraft name (if exists)
    spacecraft = re.findall(r"Dragon|Crew Dragon|Sherpa-LTC2", payload)
    spacecraft = spacecraft[0] if spacecraft else None

    return pd.Series([mission_name, satellite_count, spacecraft])


SpaceX_df[['mission_name', 'satellite_count', 'spacecraft']] = SpaceX_df["Payload"].apply(extract_payload_info)
SpaceX_df[["Payload", "mission_name", "satellite_count", "spacecraft"]].sample(20)
```

```python
pattern = r'~?([\d,]+)\s*kg\s*\(([\d,]+)\s*lb\)'
# Use .str.extract() to get the two groups (kg and lb)
SpaceX_df[['Mass_kg', 'Mass_lb']] = SpaceX_df['Payload mass'].str.extract(pattern)
```

```python
SpaceX_df[["Payload mass", "Mass_kg", "Mass_lb"]].sample(15)
```

```python
def test_mass_float_values(mass_df):
    """
    Test if the 'Mass_kg' and 'Mass_lb' columns contain only float values or NaN.
    Returns True if all values are valid floats or NaN, False otherwise.
    """
    # Check if 'Mass_kg' and 'Mass_lb' columns exist
    if 'Mass_kg' not in mass_df.columns or 'Mass_lb' not in mass_df.columns:
        print("One or both mass columns are missing.")
        return False

    # Check if all values in 'Mass_kg' are either float or NaN
    mass_kg_valid = mass_df['Mass_kg'].apply(lambda x: isinstance(x, (float, type(pd.NA))) or pd.isna(x))
    mass_lb_valid = mass_df['Mass_lb'].apply(lambda x: isinstance(x, (float, type(pd.NA))) or pd.isna(x))

    if mass_kg_valid.all() and mass_lb_valid.all():
        print("All values in 'Mass_kg' and 'Mass_lb' are valid floats or NaN.")
        return True
    else:
        if not mass_kg_valid.all():
            print("Invalid entries found in 'Mass_kg'.")
        if not mass_lb_valid.all():
            print("Invalid entries found in 'Mass_lb'.")
        return False

mass_float_result = test_mass_float_values(SpaceX_df[['Mass_kg', 'Mass_lb']])
mass_float_result
```

```python
# Define orbit simplifications
orbit_mapping = {
    r'leo iss.*|retrograde leo|leo': 'LEO',
    r'polar orbit leo|polar leo': 'Polar LEO',
    r'gto': 'GTO',
    r'sso': 'SSO',
    r'sun–earth l1.*': 'Sun-Earth L1',
    r'sun–earth l2.*': 'Sun-Earth L2',
    r'heliocentric mars transfer.*': 'Heliocentric Mars Transfer Orbit',
    r'heliocentric': 'Heliocentric',
    r'leo meo.*': 'LEO/MEO',
    r'meo': 'MEO',
    r'high elliptical heo|heo.*': 'HEO',
    r'molniya': 'Molniya',
    r'sub-orbital': 'Sub-Orbital',
    r'blt|ballistic lunar transfer': 'BLT',
    r'tli': 'TLI'
}

# Function to clean orbit names
def simplify_orbit(orbit):
    orbit = orbit.lower()
    for pattern, simplified_orbit in orbit_mapping.items():
        if re.search(pattern, orbit):
            return simplified_orbit
    return orbit.upper()  # Return unchanged if no match

# Apply the function to the 'Orbit' column in the DataFrame
SpaceX_df['simplified_orbit'] = SpaceX_df['Orbit'].apply(simplify_orbit)

# View a sample of the simplified orbit data
SpaceX_df[['Orbit', 'simplified_orbit']].sample(10)
```

```python
import re

# Define the cleaning function
def clean_booster_landing(entry):
    entry = entry.lower().strip()

    # Remove parentheses
    entry = re.sub(r'\s*\(.*?\)\s*', '', entry)

    # Standardize known patterns
    patterns = {
        r'failure\s*\(parachute\)': 'Failure - Parachute',
        r'failure\s*\(ocean\)': 'Failure - Ocean',
        r'controlled\s*\(ocean\)': 'Controlled - Ocean',
        r'failure\s*j.*?': 'Failure - JRTI',
        r'failure\s*ocisly': 'Failure - OCISLY',
        r'partial failure\s*ocisly': 'Partial Failure - OCISLY',
        r'success\s*j.*?': 'Success - JRTI',
        r'success\s*ocisly': 'Success - OCISLY',
        r'success\s*asog': 'Success - ASOG',
        r'success\s*lz\-1': 'Success - LZ-1',
        r'success\s*lz\-2': 'Success - LZ-2',
        r'success\s*lz\-4': 'Success - LZ-4',
        r'precluded\s*drone ship': 'Precluded - Drone Ship',
        r'precluded': 'Precluded'
    }

    for pattern, replacement in patterns.items():
        entry = re.sub(pattern, replacement, entry)

    # Title case for consistency
    return entry.title()

# Apply the cleaning function to the 'Booster landing' column
SpaceX_df['cleaned_booster_landing'] = SpaceX_df['Booster landing'].apply(clean_booster_landing)

# View a sample of the cleaned data
SpaceX_df[['Booster landing', 'cleaned_booster_landing']].sample(10)
```
