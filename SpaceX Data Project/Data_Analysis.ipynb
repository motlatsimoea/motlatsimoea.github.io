## List the date when the first succesful landing outcome in ground pad was acheived.

```python
%%sql select MIN(Date) AS First_Successful_Landing_Date, Launch_outcome, cleaned_booster_landing AS Landing_Outcome
      from spacex_launches
      where Launch_outcome = 'Success' and cleaned_booster_landing = 'Ocisly Success';
```

## Loading Data into a dataframe

```python
# Run the SQL query and store the result in a variable
query_result = %sql SELECT * FROM spacex_launches;

# Convert the result to a Pandas DataFrame
df = query_result.DataFrame()

# Display the DataFrame
df.head()
```

```python
df['cleaned_booster_landing'].value_counts()
```

```python
df['class'] = df['cleaned_booster_landing'].apply(lambda x: 1 if 'Success' in x else 0)
```

```python
df['Flight_No'] = pd.to_numeric(df['Flight_No'], errors='coerce')

# Optionally, drop NaN values or fill them with a specific value (e.g., 0 or the mean)
# df['Flight_No'].fillna(0, inplace=True)  # Fill NaNs with 0
# or
# df.dropna(subset=['Flight_No'], inplace=True)  # Drop rows with NaNs

# Convert to integer type, using Int64 to keep NaNs
df['Flight_No'] = df['Flight_No'].astype('Int64')
```

```python
# Filter rows where 'Flight_No' is not NaN and is <= 90
df_limited = df[df['Flight_No'].notna() & (df['Flight_No'] <= 90)]

# Plot with rotated x-axis labels to avoid overlap
sns.catplot(y="Mass_kg", x="Flight_No", hue="class", data=df_limited, aspect=5)

plt.xlabel("Flight Number", fontsize=20)
plt.ylabel("Payload Mass (kg)", fontsize=20)

# Rotate x-axis labels for better readability
plt.xticks(rotation=90, ha='right')

plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

```python
# Filter rows where 'Flight_No' is not NaN and is <= 90
df_limited = df[df['Flight_No'].notna() & (df['Flight_No'] <= 90)]

# Plot with rotated x-axis labels to avoid overlap
sns.catplot(y="Launch_site", x="Flight_No", hue="class", data=df_limited, aspect=5)

plt.xlabel("Flight Number", fontsize=20)
plt.ylabel("Launch_Site", fontsize=20)

# Rotate x-axis labels for better readability
plt.xticks(rotation=90, ha='right')

plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

```python
sns.scatterplot(y="Launch_site", x="Mass_kg", hue="class", data=df)
plt.xlabel("Payload Mass(kg)",fontsize=20)
plt.ylabel("Launch Site",fontsize=20)
plt.show()
```

```python
sns.catplot(y="simplified_orbit", x="Flight_No", hue="class", data=df_limited, aspect=5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Orbit",fontsize=20)
plt.show()
```

```python
# Plot a scatter point chart with x axis to be Payload and y axis to be the Orbit, and hue to be the class value
sns.scatterplot(y="simplified_orbit", x="Mass_kg", hue="class", data=df)
plt.xlabel("Payload (kg)",fontsize=20)
plt.ylabel("Orbit",fontsize=20)
plt.show()
```

```python
bar_df = df[['simplified_orbit', 'class']].groupby(['simplified_orbit'])['class'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='simplified_orbit', y='class', data=bar_df)
plt.xlabel("Orbit Type", fontsize=15)
plt.ylabel("Success Rate", fontsize=15)
plt.title("Success Rate by Orbit Type", fontsize=18)
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot a line chart with x axis to be the extracted year and y axis to be the success rate
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Year'] = df['Date'].dt.year
line_df = df[['Year', 'class']].groupby(['Year'])['class'].mean().reset_index()

# Create the line chart
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='class', data=line_df, marker='o')
plt.xlabel("Year", fontsize=15)
plt.ylabel("Success Rate", fontsize=15)
plt.title("Success Rate by Year", fontsize=18)
plt.grid(True)
plt.show()
```
