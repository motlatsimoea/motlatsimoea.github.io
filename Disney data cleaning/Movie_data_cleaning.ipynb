```python
import re

movie_df = df.copy()

def process_text(text):
    if isinstance(text, (str, list)):
        if isinstance(text, list):
            text = ', '.join(text)

        # Remove content inside parentheses
        text_cleaned = re.sub(r'\([^)]*\)', '', text)

        # Remove specific words and characters like 'Score', 'Background', etc.
        text_cleaned = re.sub(r'\b(Songs:|song|Score|score|Background|background|and|Songs|Song|See |plot|arrangements|Music|Composed|by|Curation|Concept|Screenplay|Dialogues|Lyrics|Live-action|Story| direction|adaptation|)\b', '', text_cleaned)

        # Remove unnecessary colons
        text_cleaned = re.sub(r'\s*:\s*', '', text_cleaned)

        # Remove stray quotes
        text_cleaned = text_cleaned.replace('"', '').replace(" ,", ",")

        # Replace '&' with a comma
        text_cleaned = re.sub(r'\s*&\s*', ', ', text_cleaned)

        # Remove unwanted periods that are not part of initials
        text_cleaned = re.sub(r'(?<!\b\w)\.\s+', '', text_cleaned)

        # Preserve initials with periods (e.g., "P. Bacon")
        text_cleaned = re.sub(r'(\b\w)\.(?!\w)', r'\1.', text_cleaned)

        # Fix cases where periods and commas are misplaced (like '.,')
        text_cleaned = re.sub(r'\.\s*,', ',', text_cleaned)
        text_cleaned = re.sub(r',\s*\.', ',', text_cleaned)

        # Replace multiple commas with a single comma
        text_cleaned = re.sub(r',+', ',', text_cleaned).strip(', ')

        # Split by commas, clean up the names, and preserve apostrophes
        parts = re.split(r',+', text_cleaned)
        names = [part.strip() for part in parts if part.strip()]

        # Join the names with a comma separator
        return ", ".join(names)

    return text
```

```python
def extract_release_date(dates):
    if isinstance(dates, list):
        dates = ', '.join(dates)

    if isinstance(dates, str):
        # Normalize spaces and remove extraneous parentheses
        dates = re.sub(r'\s+', ' ', dates)
        dates = dates.replace("(", " (").replace(")", ")")

        # Split by closing parentheses followed by optional spaces and commas
        date_location_pairs = re.findall(r'([\w\s,]+?\(\w[\w\s]+\))', dates)

        # Updated regex patterns to handle different date formats
        date_patterns = [
            r'\b\w+\s\d{1,2},\s\d{4}\b',      # Matches "Month Day, Year" format (e.g., "March 31, 2003")
            r'\d{1,2}\s\w+\s\d{4}\b',         # Matches "Day Month Year" format (e.g., "16 March 2011")
            r'\b\d{4}\b',                     # Matches "Year" format (e.g., "2020")
            r'\b\d{4}\s*–\s*\b\d{4}\b',       # Matches "Year–Year" format (e.g., "1950–present")
        ]

        # Priority order for locations
        priorities = [
            'United States',
            'Hollywood',
            'California',
            'Los Angeles',
            'United Kingdom',
            'France',
            'Spain',
            'Sundance'
        ]

        # Create a regex pattern for the priorities
        priority_pattern = re.compile(r'\(\s*({})\s*\)'.format('|'.join(re.escape(priority) for priority in priorities)), re.IGNORECASE)

        # Search through each priority and pattern combination
        for pair in date_location_pairs:
            match = priority_pattern.search(pair)
            if match:
                for pattern in date_patterns[:2]:  # Look for full dates first (month-day-year or day-month-year)
                    date_match = re.search(pattern, pair)
                    if date_match:
                        return date_match.group(0).strip()

        # Fallback: return the first valid date found if none of the priorities match
        for pattern in date_patterns:
            match = re.search(pattern, dates)
            if match:
                return match.group(0).strip()

    return None


movie_df['dates'] = movie_df['combined_release_date'].apply(extract_release_date)
movie_df[['dates', 'combined_release_date']]
```

```python
## CASES:
# $11—15 million, [$15,000,000, (1 film)], $18 million or $25 million
# $85 million, $858,000, ~$70 million, $100 million+, ₽650 million
# $12.5 million (stage production), [~$8 million, ₽, 370 million], (US$2.8 million), [$306.6 million (gross), $263.7 million (net)]
# [€57.1 million, {US$80 million)], $8 million ( ₽ 350 million), $4,000,000 (estimated), [$32 million, (estimated)]
# 60 million Norwegian Kroner (around $8.7 million in 1989), A$8.7 million, $3.5 to 4 million, AU$1 million

def extract_money_value(entry):
    def convert_to_number(value, currency='USD', multiplier=1):
        try:
            value = value.replace(',', '').replace('$', '').replace('€', '').replace('₽', '').replace('AU$', '').replace('A$', '')
            num = float(value) * multiplier
            if currency == 'RUB':  # Conversion from Ruble to USD
                num *= 0.011
            elif currency == 'AUD':  # Conversion from AUD to USD
                num *= 0.66
            elif currency == 'CRORE':  # Conversion from Crore to USD
                num *= 0.012
            return int(num)
        except ValueError:
            return None

    if isinstance(entry, (list, tuple, pd.Series, pd.DataFrame)):
        return None

    if pd.isnull(entry):
        return None

    # Convert to string if it's not already
    entry = str(entry)

    dollar_values = []
    worldwide_value = None

    # Handle values in list format, ensuring we iterate over each item in the list
    if entry.startswith('[') and entry.endswith(']'):
        entry = entry[1:-1]  # Remove the brackets

    # Updated to handle ranges like "$3.5 to 4 million"
    range_regex = re.compile(r'\$?(\d[\d.,]*\.?\d*)\s*([–—-]|to)\s*\$?(\d[\d.,]*\.?\d*)\s*(million|billion|trillion)?', re.IGNORECASE)
    matches = range_regex.findall(entry)
    for match in matches:
        value1 = float(match[0].replace(',', ''))
        value2 = float(match[2].replace(',', ''))
        multiplier = 1
        if match[3]:  # Handle 'million', 'billion', 'trillion'
            if 'million' in match[3].lower():
                multiplier = 1_000_000
            elif 'billion' in match[3].lower():
                multiplier = 1_000_000_000
        max_value = max(value1, value2) * multiplier
        dollar_values.append(int(max_value))

    # General dollar extraction (handles single amounts like "$25 million")
    if 'AU$' not in entry and 'A$' not in entry:
        dollar_regex = re.compile(r'\$([\d.,]+)(?:\s*(million|billion|trillion))?', re.IGNORECASE)
        dollar_match = dollar_regex.search(entry)

        if dollar_match:
            value = dollar_match.group(1).replace(',', '')
            multiplier = 1
            if dollar_match.group(2):
                if 'million' in dollar_match.group(2).lower():
                    multiplier = 1_000_000
                elif 'billion' in dollar_match.group(2).lower():
                    multiplier = 1_000_000_000
                elif 'trillion' in dollar_match.group(2).lower():
                    multiplier = 1_000_000_000_000
            # Convert the dollar value (assumed to be in USD) to a number
            usd_value = convert_to_number(str(float(value) * multiplier), currency='USD')
            if usd_value is not None:
                dollar_values.append(usd_value)

    # Handle Ruble values, including cases like "₽650 million" or "₽2 billion"
    ruble_regex = re.compile(r'₽([\d.,]+)\s*(million|billion)?', re.IGNORECASE)
    ruble_match = ruble_regex.search(entry)
    if ruble_match:
        value = ruble_match.group(1).replace(',', '')
        if value:  # Check if the value is not empty
            multiplier = 1
            if ruble_match.group(2):  # Check if "million" or "billion" is present
                if 'million' in ruble_match.group(2).lower():
                    multiplier = 1_000_000
                elif 'billion' in ruble_match.group(2).lower():
                    multiplier = 1_000_000_000
            # Convert the Ruble value to USD using the 0.011 conversion rate
            ruble_value = convert_to_number(str(float(value) * multiplier), currency='RUB')
            if ruble_value is not None:
                dollar_values.append(ruble_value)

    # Handle Australian Dollar values (A$ or AU$), convert to USD using 0.66 conversion rate
    aud_regex = re.compile(r'(AU\$|A\$)([\d.,]+)\s*(million|billion)?', re.IGNORECASE)
    aud_match = aud_regex.search(entry)
    if aud_match:
        value = aud_match.group(2).replace(',', '')
        multiplier = 1
        if aud_match.group(3):
            if 'million' in aud_match.group(3).lower():
                multiplier = 1_000_000
            elif 'billion' in aud_match.group(3).lower():
                multiplier = 1_000_000_000
        aud_value = convert_to_number(str(float(value) * multiplier), currency='AUD')
        if aud_value is not None and aud_value not in dollar_values:
            dollar_values.append(aud_value)

    # Handle "crore" (131 crore = 1.31 billion) - South Asian notation
    crore_regex = re.compile(r'(\d[\d.,]*)\s*(crore)', re.IGNORECASE)
    crore_match = crore_regex.search(entry)
    if crore_match:
        value = crore_match.group(1).replace(',', '')
        multiplier = 10_000_000  # 1 crore equals 10 million
        # Convert to USD using 0.012 conversion rate after multiplying by 10 million
        crore_value = convert_to_number(value, currency='CRORE', multiplier=multiplier)
        if crore_value is not None:
            dollar_values.append(crore_value)

    # Prioritize Worldwide value if mentioned
    worldwide_regex = re.compile(r'\$([\d.,]+)\s*(million|billion|trillion)?\s*\(Worldwide\)', re.IGNORECASE)
    worldwide_match = worldwide_regex.search(entry)
    if worldwide_match:
        value = worldwide_match.group(1).replace(',', '')
        multiplier = 1
        if worldwide_match.group(2):
            if 'million' in worldwide_match.group(2).lower():
                multiplier = 1_000_000
            elif 'billion' in worldwide_match.group(2).lower():
                multiplier = 1_000_000_000
        worldwide_value = convert_to_number(value, multiplier=multiplier)

    # If a worldwide value is present, prioritize it
    if worldwide_value:
        return worldwide_value

    # Handle special cases like "<$106,759,044, (Total of 2 films)>"
    special_case_regex = re.compile(r'<\$(\d[\d.,]*)\s*\(.*?\)>', re.IGNORECASE)
    special_case_match = special_case_regex.search(entry)
    if special_case_match:
        value = special_case_match.group(1).replace(',', '')
        num = convert_to_number(value)
        if num is not None:
            dollar_values.append(num)

    # Return the maximum dollar value found
    if dollar_values:
        return max(dollar_values)

    return None


# Test cases
test_cases = [
    "$11—15 million",
    "$11—15 million, [$15,000,000, (1 film)], $18 million or $25 million",
    "$85 million, $858,000, ~$70 million, $100 million+, ₽650 million",
    "$12.5 million (stage production), [~$8 million, ₽, 370 million], (US$2.8 million), [$306.6 million (gross), $263.7 million (net)]",
    "[€57.1 million, {US$80 million)], $8 million ( ₽ 350 million), $4,000,000 (estimated), [$32 million, (estimated)]",
    "60 million Norwegian Kroner (around $8.7 million in 1989)",
     "A$8.7 million",
    "$3.5 to 4 million",
    "AU$1 million",
    "131 crore",  # New case with crore
    "$15,000,000, (1 film)",
    "$35–40 million",
    "$75–90 million",
    "$150–175 million",
    "$295–387 million"
]

# Applying the function to each test case
results = [extract_money_value(case) for case in test_cases]

# Display results
for i, result in enumerate(results):
    print(f"Case {i+1}: Extracted Value: {result}")
```

```python
import pandas as pd

def convert_dates(date_str):
    try:
        # Handle formats like 'May 20, 1994' and '14 January 2011'
        return pd.to_datetime(date_str, errors='coerce', format='%B %d, %Y') or pd.to_datetime(date_str, errors='coerce', format='%d %B %Y')
    except:
        # Handle cases where only the year is present like '2022'
        return pd.to_datetime(date_str, errors='coerce', format='%Y')

# Apply the function to the movie_df['dates'] column
movie_df['dates'] = movie_df['dates'].apply(convert_dates)
movie_df[['dates']]
```

```python
movie_df[['Run Time', 'Running time']]
```

```python
def convert_run_time_to_hours_minutes(runtime_str):
    # Convert to float first, in case of decimals
    minutes = float(runtime_str)

    # Convert minutes to a pandas Timedelta object
    return pd.to_timedelta(minutes, unit='m')

# Apply the function to the 'Run Time' column
movie_df['Run Time'] = movie_df['Run Time'].apply(convert_run_time_to_hours_minutes)
movie_df[['Run Time']]
```

```python
def remove_days_from_runtime(runtime_val):
    # Ensure the value is converted to a string first
    runtime_str = str(runtime_val)

    # Split the string to remove the 'days' part and keep only the time
    if 'days' in runtime_str:
        return runtime_str.split('days')[-1].strip()
    return runtime_str  # If 'days' not found, return the original string

# Apply the function to your DataFrame column
movie_df['Run Time'] = movie_df['Run Time'].apply(remove_days_from_runtime)

# Check the result
movie_df[['Run Time']]
```

```python
movie_df[['Box office_new', 'Budget_new']].info()
```

```python
def format_monetary_value(value):
    # Format the value with commas and two decimal places
    return f"{value:,.2f}"

# Apply the function to the column with monetary values
movie_df['Box office_new'] = movie_df['Box office_new'].apply(format_monetary_value)
movie_df['Budget_new'] = movie_df['Budget_new'].apply(format_monetary_value)
# Check the result
movie_df[['Box office_new', 'Budget_new']]
```
